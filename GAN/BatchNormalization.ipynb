```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, BatchNormalization
from tensorflow.keras.models import Sequential

# Generate random inputs to simulate activations
np.random.seed(42)
x = np.random.randn(1000, 1) * 5 + 10  # Simulating activations

# Apply a simple Dense Layer (No Batch Normalization)
model_no_bn = Sequential([
    Dense(1, input_shape=(1,), use_bias=False)  # Simple transformation
])
output_no_bn = model_no_bn(x, training=True)

# Apply a Dense Layer with Batch Normalization
model_with_bn = Sequential([
    Dense(1, input_shape=(1,), use_bias=False),
    BatchNormalization()  # Normalizes activations
])
output_with_bn = model_with_bn(x, training=True)

# Plot results
plt.figure(figsize=(10, 5))

# Without Batch Normalization
plt.subplot(1, 2, 1)
plt.hist(output_no_bn.numpy(), bins=30, color='red', alpha=0.7)
plt.title("Without Batch Normalization")
plt.xlabel("Activation Value")
plt.ylabel("Frequency")

# With Batch Normalization
plt.subplot(1, 2, 2)
plt.hist(output_with_bn.numpy(), bins=30, color='blue', alpha=0.7)
plt.title("With Batch Normalization")
plt.xlabel("Activation Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
